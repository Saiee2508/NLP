```python
from transformers import pipeline

# Load sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# Function to analyze sentiment
def analyze_sentiment(text):
    result = sentiment_analyzer(text)
    return result[0]['label'], result[0]['score']

# Example usage
texts = [
    "I love this product! It's amazing.",
    "This is the worst experience I've ever had.",
    "It's an okay movie, not the best but not the worst."
]

for text in texts:
    label, score = analyze_sentiment(text)
    print(f"Text: {text}\nSentiment: {label}, Confidence: {score:.4f}\n")
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline
import torch

# Load the RoBERTa-based sentiment analysis model
MODEL_NAME = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)

# Sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Function to analyze sentiment
def analyze_sentiment(text):
    result = sentiment_analyzer(text)
    return result[0]['label'], result[0]['score']

# Example usage
texts = [
    "I absolutely love this! It's fantastic!",
    "This is the worst product I've ever used.",
    "It's okay, not the best but not the worst.",
    "I'm feeling really down today..."
]

for text in texts:
    label, score = analyze_sentiment(text)
    print(f"Text: {text}\nSentiment: {label}, Confidence: {score:.4f}\n")
```
